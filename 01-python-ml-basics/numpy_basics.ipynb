```python
# Cell 5: Dot Product â€” Linear Algebra (ML ka foundation)
x = np.array([1, 2, 3])
w = np.array([0.5, 0.3, 0.2])  # weights (ML mein model ke weights)

# Dot product = prediction nikalne ka tarika
prediction = np.dot(x, w)  # (1*0.5) + (2*0.3) + (3*0.2) = 1.7
print(f"Input: {x}")
print(f"Weights: {w}")
print(f"Prediction (dot product): {prediction}")

# Matrix multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
result = np.dot(A, B)
print(f"Matrix multiply:\n{result}")
```
