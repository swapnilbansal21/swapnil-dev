```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df["Embarked"].mode()
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df = df.drop("Cabin", axis=1)
```

```python
df
```

```python
df["Family_Size"] = df["SibSp"] + df["Parch"] + 1
```

```python
df
```

```python
df['Is_Alone']= (df['Family_Size'] == 1).astype(int)
```

```python
df.head()
```

```python
fig,axes =plt.subplots(1,3,figsize=(15,4))
```

```python
sns.countplot(x="Survived", hue="Survived", data=df,
              palette=["#ff6b6b", "#51cf66"], legend=False, ax=axes[0])
```

```python
# Bar 1:
sns.countplot(x="Survived", hue="Survived", data=df,
              palette=["#ff6b6b", "#51cf66"], legend=False, ax=axes[0])

# Bar 2:
sns.countplot(x="Pclass", hue="Pclass", data=df,
              palette="Set2", legend=False, ax=axes[1])

# Bar 3:
sns.countplot(x="Sex", hue="Sex", data=df,
              palette=["#339af0", "#ff6b6b"], legend=False, ax=axes[2])
```

```python
fig, axes = plt.subplots(1, 6, figsize=(15, 4))
# Ye line ZARURI hai — bina iske axes exist nahi karta

# Bar 1:
sns.countplot(x="Survived", hue="Survived", data=df,
              palette=["#ff6b6b", "#51cf66"], legend=False, ax=axes[0])
axes[0].set_title("Survived vs Died")
axes[0].set_xticklabels(["Died", "Survived"])

# Bar 2:
sns.countplot(x="Pclass", hue="Pclass", data=df,
              palette="Set2", legend=False, ax=axes[1])
axes[1].set_title("Passenger Class")

# Bar 3:
sns.countplot(x="Sex", hue="Sex", data=df,
              palette=["#339af0", "#ff6b6b"], legend=False, ax=axes[2])
axes[2].set_title("Gender Distribution")

plt.tight_layout()
plt.show()
```

```python
# Cell 3: countplot with HUE — Gender wise survival
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.countplot(x="Sex", hue="Survived", data=df,
              palette=["#ff6b6b", "#51cf66"], ax=axes[0])
axes[0].set_title("Survival by Gender")
axes[0].legend(["Died", "Survived"])

sns.countplot(x="Pclass", hue="Survived", data=df,
              palette=["#ff6b6b", "#51cf66"], ax=axes[1])
axes[1].set_title("Survival by Class")
axes[1].set_xticklabels(["1st", "2nd", "3rd"])
axes[1].legend(["Died", "Survived"])

plt.tight_layout()
plt.show()
```

```python
fig, axes =plt.subplots(1,2,figsize=(15,4))
sns.countplot(x="Sex", hue="Survived", data=df,palette=["#ff6b6b", "#51cf66"], ax=axes[0])
axes[0].set_title("Survived vs Died")
axes[0].legend(["Died", "Survived"])
```

```python
fig,axes =plt.subplots(1,2,figsize=(15,4))
sns.histplot(df['Age'],bins =30,kde =True,ax=axes[0])
axes[0].set_title("Age distribution")
axes[0].axvline(df["Age"].mean(), color="red", linestyle="--",
                label=f'Mean: {df["Age"].mean():.1f}')
sns.histplot(data=df, x="Age", hue="Survived", bins=30,
             palette=["#ff6b6b", "#51cf66"], ax=axes[1])
axes[1].set_title("Age Distribution by Survival")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.boxplot(x="Pclass", y="Age", data=df,
            hue="Pclass", palette="Set2", legend=False, ax=axes[0])
axes[0].set_title("Age Distribution by Class")
axes[0].set_xticklabels(["1st", "2nd", "3rd"])

sns.boxplot(x="Pclass", y="Fare", data=df,
            hue="Pclass", palette="Set2", legend=False, ax=axes[1])
axes[1].set_title("Fare Distribution by Class")
axes[1].set_xticklabels(["1st", "2nd", "3rd"])

plt.tight_layout()
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="RdYlBu",
            center=0, fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()

print("\nCorrelation with Survived:")
print(numeric_df.corr()["Survived"].sort_values(ascending=False))
```
